```python
def Outdet(data):
    Q1=data.quantile(0.25)
    Q3=data.quantile(0.75)
    IQR=Q3-Q1
    LR=Q1-(IQR*1.5)
    UR=Q3+(IQR*1.5)
    return Q1,Q3,IQR,LR,UR
```

```python
Q1,Q3,IQR,LR,UR=Outdet(outcome_zero.Pregnancies)
print(Q1,Q3,IQR,LR,UR)
```

```python
outcome_zero=outcome_zero[(outcome_zero['Pregnancies'] > LR)  &  (outcome_zero['Pregnancies'] < UR)]
outcome_zero
```
